```python
def calcFatorial(x: int):
    resultado = 1
    for i in range(1,x+1):
        resultado *= i
    return resultado

def calcFatorialRec(n: int):
    if (n == 0):
        return 1
    return n * calcFatorial(n-1)
```

### Operações com Vetores

Os vetores serão indicados por *letras minúsculas em negrito*.

#### Adição

${\mathbf a} + {\mathbf b} = (a_1, a_2, \cdots, a_n) + (b_1, b_2, \cdots, b_n) = (a_1 + b_1, a_2 + b_2, \cdots, a_n + b_n)$

${\mathbf a} = (p_x, p_y, p_z)$

$(5, 0, 2) + (1, 6, 4) = (5+1, 0+6, 2+4)$

#### Multiplicação de vetor por escalar

$k {\mathbf a} = k (a_1, a_2, \cdots, a_n) = (k a_1, k a_2, \cdots, k a_n)$

$2 (3, -2, 5, 1, -8) = (6, -4, 10, 2, -16)$

$3 (2, 1, 5) = (6, 3, 15)$

$0.5 (2, 10, 5) = (1, 5, 2.5)$

```python
# Pista de skate: Alexandre Luiz

# Função definidas por partes
def tampa(x):
    if x < -2 and x > -4:
        return 6
    if x < 4 and x > 2:
        return 6
    
def caixaSemTampa(x):
    if x < 3 and x!=-3:
        return 2
    elif x==3 and x<4:
        return 2*x
    elif x==-3 and x>-4:
        return -2*x
    
X = np.linspace(-3, 3, 100)
z = np.linspace(-2, 2, 100)
y = np.linspace(-3, 3, 100)


Fm = [caixaSemTampa(x) for x in X]
Jm = [tampa(x) for x in y]
parabola = z**2+2

plt.plot(X, Fm, color="green")
plt.plot(z, parabola, color="black")
plt.plot(y, Jm, color="green")
plt.title("Pista de Skate do Alê")

plt.show()
```

#### Operações vetoriais

```python
somaV = lambda v1,v2: tuple([(i+j) for i,j in zip(v1,v2)])
prod_por_escalar = lambda k,v: tuple([k*i for i in v])

def diferencaVetores(v1, v2):
    assert len(v1) == len(v2), "Os vetores precisam ter mesma dimensão"
    return tuple(v1[i] - v2[i] for i in range(len(v1)))
```

#### Reta que passa por $p_0$ e é paralela a $t$: *equação vetorial*

`p(u) = p0 + u*t`

```python
# reta vetorial
p = lambda p, t: lambda u: somaV(p, prod_por_escalar(u, t))
```
